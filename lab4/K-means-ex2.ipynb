```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
plt.figure(figsize=(12,12))
```

```python
# initial setting
n_samples = 600
random_state = 170
n_clusters = 7
```

```python
X, y = make_blobs(n_samples=n_samples, random_state=random_state, centers=n_clusters)

# start --> 10행 2열 
start = 20*np.random.rand(n_clusters, 2) - 10
```

```python
start
```

```python
def kplot(centroids, title, c=None):
    plt.scatter(X[:,0], X[:,1], c=c)
    plt.scatter(centroids[:, 0], centroids[:,1], color='black')
    
    # 행의 갯수 (클러스터의 갯수) 만큼 iteration 하면서 
    for i in range(centroids.shape[0]):
        plt.text(centroids[i,0], centroids[i,1], str(i), fontsize=20)
    plt.title(title)
```

```python
def runkmeans(niter):
    # n_init --> Number of time the k-means algorithm will be run with different centroid seeds. 
    # The final results will be the best output of n_init consecutive runs in terms of inertia.
    # inertia --> a tendency to do nothing or to remain unchanged.
    km = KMeans(n_clusters=n_clusters, n_init = 1, max_iter=niter, init=start)
    
    y_pred = km.fit_predict(X)
    centroids = km.cluster_centers_
    
    # y 의 prediction 이랑 centroids 를 넘겨주면 된다. 
    return y_pred, centroids
```

```python
plt.subplot()
kplot(start, 'Original data')
```

```python
y_pred, centroids = runkmeans(1)
kplot(centroids, '1st iteration', c=y_pred)
```

```python
y_pred, centroids = runkmeans(2)
kplot(centroids,'2nd iterations',c=y_pred)
```

```python
y_pred, centroids = runkmeans(1000)
kplot(centroids,'1000th iterations',c=y_pred)
```
