```python
from sklearn.datasets import fetch_openml
from sklearn.preprocessing import StandardScaler, scale 
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
def plot_k_means(X, cluster_assignments, centroid_locations):
    plt.figure(figsize=(6, 6))
    plt.viridis() # Set colour map
    plt.scatter(X[:, 0], X[:, 1], s=20, c=cluster_assignments, alpha=0.8) # plot data points
    plt.scatter(centroid_locations[:, 0], centroid_locations[:, 1], s=200, marker='X', c=range(10), edgecolors='k') # plot centroids
    plt.show()
```

```python
mnist = fetch_openml(name='mnist_784')
data = mnist.data
y = mnist.target
```

```python
pca = PCA(n_components=2)
X_scaled = StandardScaler().fit_transform(data)
X = pca.fit_transform(X_scaled)
```

```python
K_values = range(1, 15)
scores = []

for K in K_values:
    kmeans = KMeans(K, random_state=0).fit(X)
    scores.append(-kmeans.score(X))
    
plt.figure()
plt.plot(K_values, scores, label='error curve')
plt.xlabel('Cluster numbers')
plt.ylabel('Error rate')
plt.legend(loc='best')
plt.show()
```

```python
y_pred = KMeans(n_clusters=10, random_state=0).fit_predict(X)
```

```python
new_data_with_labels = np.vstack((X.T, y_pred)).T
df = pd.DataFrame(data=new_data_with_labels, columns=('1st PC', '2nd PC', 'labels'))
```

```python
sns.FacetGrid(df, hue='labels', height=10).map(plt.scatter, '1st PC', '2nd PC').add_legend()
plt.show()
```
